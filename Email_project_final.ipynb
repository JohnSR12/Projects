```python
import csv
import re

def register():
    with open("email_registration.csv",mode="a",newline="") as f:
        writer = csv.writer(f,delimiter=",")
        condition1 = "^[a-z]+[a-z 0-9 . _]+[@]\w+[.]\w{2,3}$"
        email = input("Enter your email : ")
        a = 0
        b = 0
        if re.search(condition1,email):
            a = 1
            print("Valid email")
        else:
            a = 0
            print("Invalid email")
        if a == 1:
            condition2 = "^(?=.*?[0-9])(?=.*?[A-Z])(?=.*?[a-z])(?=.*[@$!%*#?&])[A-Za-z\d@$!#%*?&]{6,15}$"
            print("\nPassword Conditions \n length >5 to <16")
            print(" one special character \n one digit \n one uppercase \n one lowercase ")
            password = input("\nEnter your password : ")
            if re.search(condition2,password):
                b = 1
                print("Valid password:")
            else:
                print("Invalid password")
        if b == 1:
            writer.writerow([email,password])
            print("\n Registration is successful!")
        else:
            print("\n Please try again.")

def login():
    email = input("Please enter mail: ")
    with open("email_registration.csv", mode="r") as f:
        reader = csv.reader(f)
        found=0
        for row in reader:
            if row[0] == email:
                password = input("Please enter password: ")
                if row[1] == password:
                    print("You are logged in!")
                    return True
                else:
                    print("Invalid password, Please try agian")
                    return False
        print("Email not found, Please Register!")
        return False

def forgot():
    email = input("Enter email: ")
    with open("email_registration.csv", mode="r") as f:
        reader = csv.reader(f)
        found=0
        for i in reader:
            if i[0] == email:
                print("Password is:",i[1])
                found=1
        f.close()
        if found == 0:
            print("Email id not found, pls register")

option = (input("Enter 1 to Register\nEnter 2 to Login\nEnter 3 to Retrieve password: "))
if option == "1":
    print("\nRegistration Menu")
    register()
elif option == "2":
    print("\nLogin Menu")
    login()
elif option == "3":
    print("\nForgot password menu")
    forgot()
else:
    print("Invalid option, try again")
```
